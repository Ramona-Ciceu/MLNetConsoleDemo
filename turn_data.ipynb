```python
import csv
import random

# Board spaces
luck_spaces = [3, 7, 13, 17, 23, 27, 33, 37, 43, 47, 53, 57]
credit_spaces = [5, 15, 25, 35, 45, 55]
roll_again_spaces = [10, 30, 50]
center_path_spaces = [20, 40, 60]
final_center_path = list(range(61, 71))

# Luck cards
luck_cards = [
    "Powerful: Advance to Start of any 3 Paths, If Already There, Advance to Infinity",
    "Banished: Go to Odds Lock",
    "Banished: Go to Evens Lock",
    "Take Another Turn (It's Your Lucky Day)",
    "Superstar! Take Another Turn for Sharing Your Skills",
    "Awesomeness! Take Another Turn to Show Your Brilliant Educational Skills",
    "Caught in the Act! Lose a Turn",
    "Wrong Change! Lose 2 Turns",
    "Oops! Lose School Bag, Pay 50 Credits or Lose 2 Turns",
    "Move 3 Spaces Forward",
    "Move 3 Spaces Forward",
    "Move 5 Spaces Forward",
    "Move 5 Spaces Forward",
    "Move 7 Spaces Forward",
    "Move 7 Spaces Forward",
    "Winner! Receive 2 Money Tokens",
    "Good Citizen! Receive 1 Money Token for Charity Work",
    "Eureka! Win Math Quiz, Collect 2 Money Tokens",
    "Games Arcade! Receive Token from Cash",
    "It's ShowTime! Reveal Money Tokens, Give Back a Token if Over 50 Credits",
    "Let It Go! Give Credit Tokens Back to the Cash Vault"
]

luck_used_options = ["Yes", "No"]
lock_options = ["Yes", "No"]
opponent_nearby_options = ["Yes", "No"]
# Function to simulate move and decide best operation and direction
def decide_best_move(dice1, dice2, credits, position, luck_used, in_lock, opponent_nearby):
    moves = {
        "Add": position + (dice1 + dice2),
        "Subtract": max(0, position - abs(dice1 - dice2)),
        "Multiply": position + (dice1 * dice2),
        "Divide": position + (dice1 // dice2) if dice2 != 0 and dice1 % dice2 == 0 else position + abs(dice1 - dice2)
    }
    
    move_spaces = {}
    for move, target_pos in moves.items():
        if target_pos in luck_spaces:
            move_spaces[move] = "Luck"
        elif target_pos in credit_spaces:
            move_spaces[move] = "Credit"
        elif target_pos in roll_again_spaces:
            move_spaces[move] = "RollAgain"
        elif target_pos in center_path_spaces:
            move_spaces[move] = "CenterPath"
        elif target_pos in final_center_path:
            move_spaces[move] = "FinalCenter"
        else:
            move_spaces[move] = "Neutral"

    if in_lock == "Yes":
        return "Add"
    if credits >= 190:
        if any(v in ("CenterPath", "FinalCenter") for v in move_spaces.values()):
            return next(k for k, v in move_spaces.items() if v in ("CenterPath", "FinalCenter"))
        else:
            return "Multiply"
    if "Credit" in move_spaces.values():
        return next(k for k, v in move_spaces.items() if v == "Credit")
    if "Luck" in move_spaces.values() and random.random() < 0.3:
        return next(k for k, v in move_spaces.items() if v == "Luck")
    if "RollAgain" in move_spaces.values():
        return next(k for k, v in move_spaces.items() if v == "RollAgain")

    return random.choice(["Add", "Multiply"])

# Now generate the dataset
with open('turn_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Dice_Roll_1", "Dice_Roll_2", "Credits_Before", "Position_Before", 
                     "Luck_Card_Used", "In_Lock", "Opponent_Close", "Move_Direction", "Best_Move_Operation"])

    for _ in range(1500):
        dice1 = random.randint(1, 12)
        dice2 = random.randint(1, 12)
        credits = random.randint(0, 250)
        position = random.randint(0, 70)
        luck_used = random.choice(["Yes", "No"])
        in_lock = random.choice(["Yes", "No"])
        opponent_nearby = random.choice(["Yes", "No"])

        best_move = decide_best_move(dice1, dice2, credits, position, luck_used, in_lock, opponent_nearby)

        # Decide Move Direction based on operation
        if best_move in ["Add", "Multiply", "Divide"]:
            move_direction = "Forward"
        else:
            move_direction = "Backward"

        writer.writerow([dice1, dice2, credits, position, luck_used, in_lock, opponent_nearby, move_direction, best_move])

print("New dataset generated with Move Direction: turn_data.csv ✅")
```
